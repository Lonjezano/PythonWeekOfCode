```python
import pandas as pd
import requests
import html5lib
from bs4 import BeautifulSoup
```

```python
title = []
posted = []
deadline = []

# For loop to loop through all 7 pages
for i in range(1,7):
    url = f"http://hec.ac.mw/scholarships/{i}" if i > 1 else "http://hec.ac.mw/scholarships"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html5lib')
    block = soup.find_all("div" ,class_ = 'scholar-content-box')
    
    # for loop to loop each scholarship block
    for i in block:
        # For title
        title_holder = i.find("h4").text.strip()
        title.append(title_holder)
        #--------------------------------------------------------------------------------------------

        # For posted date
        posted_holder = i.find(class_="scholar-meta")
        posted_date_temp = posted_holder.text.split()

        # if i was working in my usual IDE i would have created seperate function for date extraction
        posted_date = f"{posted_date_temp[2]} {posted_date_temp[3]} {posted_date_temp[4]}"
        posted.append(posted_date)
        #------------------------------------------------------------------------------------------

        # For the deadline date
        deadline_holder = posted_holder.find_next("p").text
        deadline_date_temp = deadline_holder.split()

        # if i was working in my usual IDE i would have created seperate function for date extraction
        deadline_date = f"{deadline_date_temp[-3]} {deadline_date_temp[-2]} {deadline_date_temp[-1]}"
        deadline.append(deadline_date)

        #---------------------------------------------------------------------------------------------
    

    
    
```

```python
df = pd.DataFrame({'Title':pd.Series(title), 'Date posted':pd.Series(posted), 'Closing date':pd.Series(deadline)})
df
```
